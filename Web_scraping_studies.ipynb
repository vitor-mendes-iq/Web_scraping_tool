```python
# librarys
from tqdm import tqdm
from bs4 import BeautifulSoup
import codecs
import seaborn
import time
import urllib
from urllib.request import Request, urlopen
```

```python
# Function to define path elementos of looping
def pathFinder(n,f):
    n = n
    f = f
    paths = list()
    for i in range(0,n):
        str_n = str(i)
        file = str_n + f
        paths.append(file) 
    return(paths)
```

```python
# All text tags
with codecs.open('0.html', 'r', encoding='utf-8',
                 errors='ignore') as fdata:
        contend = fdata.read()
        soup = BeautifulSoup(contend, 'lxml')
        tags = soup.find_all('span') # find tags
        for tag in tags:
            print(tag.text)
```
