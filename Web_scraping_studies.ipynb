```python
# librarys
from bs4 import BeautifulSoup
import codecs
import seaborn
import time
import urllib
from urllib.request import Request, urlopen
```

```python
# Function to define path elementos of looping
def pathFinder(n,f):
    n = n
    f = f
    paths = list()
    for i in range(0,n):
        str_n = str(i)
        file = str_n + f
        paths.append(file) 
    return(paths)
```

```python
# All text tags
with codecs.open('0.html', 'r', encoding='utf-8',
                 errors='ignore') as fdata:
        contend = fdata.read()
        soup = BeautifulSoup(contend, 'lxml')
        tags = soup.find_all('span') # find tags
        for tag in tags:
            print(tag.text)
```

```python
# variables
path = pathFinder(50,'.html')
tab = []
refreshrate = "20"
refreshrate = int(refreshrate)

#download and name HTML web pages
for p in path:
    time.sleep(refreshrate)
    req = Request('https://br.investing.com/crypto/bitcoin/bitcoin-futures', headers={'User-Agent': 'Mozilla/5.0'})
    webpage = urlopen(req).read()
    f = open(p, 'wb')
    f.write(webpage)
    f.close

# extract information value
for pag in path:
    # Import HTML file and encoding
    with codecs.open(pag, 'r', encoding='utf-8',
                 errors='ignore') as fdata:
        contend = fdata.read()
        soup = BeautifulSoup(contend, 'lxml')
        tags = soup.find_all('span') # find tags
        # scraping Bitcoin value
    try:
        bsb_value = tags[78].text #objeticve tag
        bsb_value = float(bsb_value)
        tab.append(bsb_value)
    except ValueError:
        bsb_value = bsb_value[:-2]
        tab.append(bsb_value)
x = list(range(0,len(tab)))
seaborn.lineplot(x,tab)
```
